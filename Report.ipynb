# Informe de Proyecto: 

### ¿Qué pasos del plan se realizaron y qué pasos se omitieron (explica por qué)?

**Pasos Realizados:**

1. **Análisis Exploratorio de Datos (EDA)**: Se llevó a cabo un análisis detallado de las características de los datos, incluyendo distribuciones, correlaciones y análisis de valores faltantes.


2. **Preprocesamiento de Datos**: Se realizó la limpieza, transformación y normalización de los datos, incluyendo el manejo de valores nulos y la codificación de variables categóricas.


3. **División de Datos**: Los datos fueron divididos en conjuntos de entrenamiento, validación y prueba para garantizar una evaluación justa de los modelos.


4. **Entrenamiento de Modelos**: Se entrenaron múltiples modelos de clasificación, como Logistic Regression, Decision Tree, Random Forest, CatBoost, Gradient Boosting y LGBM Classifier, con ajustes de hiperparámetros a través de `GridSearchCV`.


5. **Evaluación de Modelos**: Se evaluaron los modelos con las métricas de `AUC-ROC`, `accuracy` y `F1-score`.

**Pasos Omitidos:**

- **Ingeniería de Características Avanzada**: Aunque se realizaron algunos ajustes básicos, no se desarrollaron características adicionales más complejas debido a limitaciones de tiempo.


- **Validación Cruzada Extensiva**: Se optó por una validación simple de tres pliegues (`cv=3`) para reducir el tiempo de cómputo, omitiendo validaciones más exhaustivas.

### ¿Qué dificultades encontraste y cómo lograste resolverlas?

**Dificultades Encontradas:**

1. **Desbalance de Clases**: Los datos presentaban un desbalance de clases que podía afectar la precisión de los modelos.

   - **Solución**: Se aplicaron técnicas de muestreo, como el sobremuestreo de la clase minoritaria, para equilibrar el conjunto de datos.


2. **Tiempo de Entrenamiento**: Algunos modelos, como `LGBM Classifier`, requirieron tiempos de entrenamiento extensos.

   - **Solución**: Se optimizaron los hiperparámetros con un enfoque de `GridSearchCV` más limitado y se utilizaron opciones paralelas (`n_jobs=-1`) para acelerar el proceso.


3. **Sobreajuste en Modelos Complejos**: Modelos como `Decision Tree` mostraban un rendimiento significativamente mejor en el conjunto de entrenamiento en comparación con el de validación.

   - **Solución**: Se limitaron los parámetros como `max_depth` y `min_samples_split` para reducir el sobreajuste.

### ¿Cuáles fueron algunos de los pasos clave para resolver la tarea?

- **Análisis Exploratorio de Datos (EDA)**: Entender la distribución de las características y la proporción de clases fue esencial para definir estrategias de preprocesamiento.


- **Preprocesamiento Completo**: Incluir codificación adecuada de variables categóricas y manejo de valores faltantes fue crucial para mejorar la calidad del modelo.


- **Optimización de Hiperparámetros**: El uso de `GridSearchCV` permitió encontrar configuraciones óptimas para cada modelo, mejorando así la precisión y el `AUC-ROC`.


- **Evaluación Multimétrica**: Evaluar los modelos con múltiples métricas (AUC-ROC, `accuracy`, `F1-score`) proporcionó una visión integral de su rendimiento.

### ¿Cuál es tu modelo final y qué nivel de calidad tiene?

**Modelo Final**: El modelo seleccionado para la implementación es el **Random Forest** debido a su equilibrio entre precisión y generalización. 

**Nivel de Calidad**:

- **AUC-ROC en Validación**: 0.8294

- **AUC-ROC en Prueba**: 0.8108

- **Tiempo de Entrenamiento**: 60 s

- **Tiempo de Predicción**: 0.25 s

Este modelo ofrece un rendimiento robusto con un `AUC-ROC` competitivo y una precisión razonable en todos los conjuntos. Es capaz de manejar grandes volúmenes de datos con un tiempo de predicción rápido, lo que lo hace adecuado para aplicaciones en producción.
